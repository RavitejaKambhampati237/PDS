B. REMOVING THE ATTRIBUTES

```python
df['Mileage'] = df['Mileage'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df['Engine'] = df['Engine'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df['Power'] = df['Power'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df['New_Price'] = df['New_Price'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
df.head()
```

```python
df.isnull().sum()
```

A. MISSING VALUES (MEAN, MEDIAN, MODE)

```python
df['Mileage'].fillna(df['Mileage'].median(),inplace=True)
df['Engine'].fillna(df['Engine'].median(),inplace=True)
df['Power'].fillna(df['Power'].median(),inplace=True)

df['Seats'].fillna(df['Seats'].mode()[0],inplace=True)

df.drop('New_Price', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

We are replacing any missing figures with mean values in the power, engine, and mileage sections. The reason for this is that the mean compares the majority of the numbers, and outliers have little effect on the mean. We use the mode to fill in any blank values in the Seat column because it represents the number of seats that occur most frequently. Since over half of the New Price column's values are missing, the entire column is being eliminated to avoid inaccurate data and analysis.

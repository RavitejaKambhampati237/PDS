```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data=pd.read_csv('/content/StudentsPerformance.csv')
```

```python
import os
import matplotlib.pyplot as plt

# Check if the directory exists, and if not, create it
output_dir = '/content/student_performance/results/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Generate your plot (assuming the plot code is correct)
plt.bar(data['gender'].value_counts().index, data['gender'].value_counts().values)
plt.title('Gender Distribution')
plt.xticks(data['gender'].value_counts().index, ['Female', 'Male'])

# Save the figure in the directory
plt.savefig(os.path.join(output_dir, 'genderdistribution.png'))

# Show the plot
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the column 'Total score' exists by summing the individual scores
if 'Total score' not in data.columns:
    data['Total score'] = data['math score'] + data['reading score'] + data['writing score']

# Create the directory if it doesn't exist
output_dir = '/content/student_performance/results/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Creating the subplots for different score distributions based on gender
figure, axis = plt.subplots(2, 2, figsize=(10, 8))

# Boxplots for each subject's score based on gender
sns.boxplot(x=data['gender'], y=data['math score'], ax=axis[0, 0])
axis[0, 0].set_title('Math Score by Gender')

sns.boxplot(x=data['gender'], y=data['reading score'], ax=axis[0, 1])
axis[0, 1].set_title('Reading Score by Gender')

sns.boxplot(x=data['gender'], y=data['writing score'], ax=axis[1, 0])
axis[1, 0].set_title('Writing Score by Gender')

sns.boxplot(x=data['gender'], y=data['Total score'], ax=axis[1, 1])
axis[1, 1].set_title('Total Score by Gender')

# Adjust layout
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
plt.suptitle('Scores Distribution by Gender')

# Save the figure
plt.savefig(os.path.join(output_dir, 'scoresdistribution.png'))

# Show the plot
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the column 'Total score' exists by summing the individual scores
if 'Total score' not in data.columns:
    data['Total score'] = data['math score'] + data['reading score'] + data['writing score']

# Map gender to colors
gender_color_map = {'male': 'blue', 'female': 'pink'}
data['gender_color'] = data['gender'].map(gender_color_map)

# Create the directory if it doesn't exist
output_dir = '/content/student_performance/results/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Creating the subplots for different score distributions based on gender
figure, axis = plt.subplots(2, 2, figsize=(10, 8))

# Boxplots for each subject's score based on gender
sns.boxplot(x=data['gender'], y=data['math score'], palette=gender_color_map, ax=axis[0, 0])
axis[0, 0].set_title('Math Score by Gender')

sns.boxplot(x=data['gender'], y=data['reading score'], palette=gender_color_map, ax=axis[0, 1])
axis[0, 1].set_title('Reading Score by Gender')

sns.boxplot(x=data['gender'], y=data['writing score'], palette=gender_color_map, ax=axis[1, 0])
axis[1, 0].set_title('Writing Score by Gender')

sns.boxplot(x=data['gender'], y=data['Total score'], palette=gender_color_map, ax=axis[1, 1])
axis[1, 1].set_title('Total Score by Gender')

# Adjust layout
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
plt.suptitle('Scores Distribution by Gender')

# Save the figure
plt.savefig(os.path.join(output_dir, 'scoresdistribution.png'))

# Show the plot
plt.show()
```

```python
#visualisation 4

figure, axis = plt.subplots(2,2)

sns.histplot(data['math score'],ax=axis[0,0],label='math score')
sns.histplot(data['reading score'],ax=axis[0,1],label='reading score')
sns.histplot(data['writing score'],ax=axis[1,0],label='writing score')
sns.histplot(data['Total score'],ax=axis[1,1],label='Total score')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Frequency distribution of scores')
plt.savefig('/content/student_performance/results/scoresfrequency.png')
```
